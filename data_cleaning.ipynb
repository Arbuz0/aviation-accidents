```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
con = sqlite3.connect('airplanes.db')
data_import = pd.read_sql_query('SELECT * from airplane_accidents', con)

data_import
```

Data issues:  
-inconsistent date format  
-leading spaces in few columns  
-redundant columns  
-aboard and fatalities column need to be split into 3 separate ones  

```python
data_import = data_import.replace('?', np.nan)
```

```python
data_import['location'] = data_import['location'].str.strip()
data_import['operator'] = data_import['operator'].str.strip()
data_import['route'] = data_import['route'].str.strip()
data_import['ac_type'] = data_import['ac_type'].str.strip()
data_import['registration'] = data_import['registration'].str.strip()
data_import
```

```python
data_import[['aboard_total', 'aboard_passengers', 'aboard_crew']] = data_import['aboard'].str.split(';', expand=True)

data_import['aboard_passengers'] = (data_import['aboard_passengers'].str)[12:]
data_import['aboard_crew'] = (data_import['aboard_crew'].str)[6:]

data_import = data_import.replace('?', np.nan)

data_import = data_import.astype({'aboard_total': float, 'aboard_passengers': float, 'aboard_crew': float}, errors='raise')

data_import[['aboard_total', 'aboard_passengers', 'aboard_crew']]
```

```python
data_import[['fatalities_total', 'fatalities_passengers', 'fatalities_crew']] = data_import['fatalities'].str.split(';', expand=True)

data_import['fatalities_passengers'] = (data_import['fatalities_passengers'].str)[12:]
data_import['fatalities_crew'] = (data_import['fatalities_crew'].str)[6:]

data_import = data_import.replace('?', np.nan)

data_import = data_import.astype({'fatalities_total': float, 'fatalities_passengers': float, 'fatalities_crew': float}, errors='raise')

data_import[['fatalities_total', 'fatalities_passengers', 'fatalities_crew']]
```

```python
data_import = data_import.drop({'aboard', 'fatalities'}, axis=1)
data_import.head()
```

```python
data_import['date'] = pd.to_datetime(data_import['date'], format='%B %d, %Y')
data_import['date'] = data_import['date'].dt.date
```

```python
data_import.dtypes
```

```python
def fix_time(time):
    if time is np.nan:
        return time
    
    if ':' in str(time):
        index = time.rfind(':')
        minutes = int(time[index + 1:index + 3])
        hours = int(time[max(index - 2, 0):index])
    else:
        minutes = int(time) % 100
        hours = int((int(time) - minutes) / 100)

    return f'{hours},{minutes}'


data_import['time'] = data_import['time'].replace('06;15', '06:15')
data_import['time'] = data_import['time'].replace('0500Z', '0500')

data_import['time'] = data_import['time'].apply(fix_time)

data_import[['hours', 'minutes']] = data_import['time'].str.split(',', expand=True)

data_import = data_import.drop('time', axis=1)

data_import.head()
```
